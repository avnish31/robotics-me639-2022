```python
from sympy.physics.mechanics import dynamicsymbols
```

```python
theta1,theta2,l1,l2 = dynamicsymbols('theta1,theta2,l1,l2')
theta1,theta2,l1,l2
```

```python
px =l1*sp.cos(theta1) +l2*sp.cos(theta1+theta2) #tip postion in x-direc
py =l1*sp.sin(theta1) +l2*sp.sin(theta1+theta2) #tip postion in y-direc
```

```python
a11=sp.diff(px,theta1)   #differentiate px with respect to theta1
a12=sp.diff(px,theta2)   #differentiate px with respect to theta2

a21=sp.diff(py,theta1)   #differentiate py with respect to theta1
a22=sp.diff(py,theta2)   #differentiate py with respect to theta2
```

```python
J =sp.Matrix([[a11 , a12],[a21 , a22]])
Jsim = sp.simplify(J)
Jsim
```
